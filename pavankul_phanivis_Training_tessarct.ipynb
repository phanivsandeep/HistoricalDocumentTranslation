```python
import os
import subprocess
```

```python
tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
image_folder = r'Images'
for filename in os.listdir(image_folder):
    if filename.endswith(".jpg"):
        image_path = os.path.join(image_folder, filename)
        print(f"Processing {image_path}")
        output_base = os.path.splitext(image_path)[0]

        # Create the box file command
        command = [tesseract_cmd, image_path, output_base, '-c', 'tessedit_create_boxfile=1']

        # Run the command and check for errors
        try:
            result = subprocess.run(command, check=True)
            if result.returncode == 0:
                print(f"Box file created for {filename}")
        except subprocess.CalledProcessError as e:
            print(f"Failed to create box file for {filename}: {e}")
```

```python
import os
import subprocess
```

```python
tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
image_folder = r'Images'

for filename in os.listdir(image_folder):
    if filename.endswith(".jpg"):
        image_path = os.path.join(image_folder, filename)
        output_base = os.path.splitext(image_path)[0]
        command = [tesseract_cmd, image_path, output_base, 'nobatch', 'box.train']
        try:
            result = subprocess.run(command, check=True)
            if result.returncode == 0:
                print(f"Training file created for {filename}")
        except subprocess.CalledProcessError as e:
            print(f"Failed to create training file for {filename}: {e}")
```
